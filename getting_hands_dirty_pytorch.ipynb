```python
## math operations
```

```python
#adding

tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,4,9,6])


tensor_a + tensor_b
```

```python
#adding 2

torch.add (tensor_a, tensor_b)
```

```python
#subtract

tensor_a - tensor_b
```

```python
#subtract 2

torch.sub(tensor_a, tensor_b)
```

```python
#multiply
tensor_a * tensor_b
```

```python
#multiply 2

torch.mul(tensor_a, tensor_b)
```

```python
#divide

tensor_a / tensor_b
```

```python
#divide 2

torch.div(tensor_a, tensor_b)
```

```python
#remainders

tensor_a % tensor_b
```

```python
#remainders 2

torch.remainder(tensor_a, tensor_b)
```

```python
#power & exponent

torch.pow(tensor_a, tensor_b)
```

```python
#reassignment_ # use underscore after every or any torch.operatior

tensor_a.add_(tensor_b)

#making tensor_a = tensor_a + tensor_b
```
